```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# グラフのサイズの指定
plt.figure(figsize=(4,4))
# 引数に指定された値のプロット。引数として、描画する点のx座標, y座標の設定、凡例に表示する文字列の設定
plt.plot([0,2,4,6,8], [0,6,12,24,48],  label="graph")
# x, y軸のラベルの設定
plt.xlabel("x")
plt.ylabel("y")
# タイトルの設定
plt.title("title")
#　凡例(左上のやつ)の表示
plt.legend()
# グラフの出力
plt.show()
```

```python
# 演習1
x = np.linspace(0, 10, 20)
y = x**2
z = (x-1)**3
w = ((x-2)*(x-4))**2
plt.plot(x, y, label="y")
plt.plot(x, z, label="z")
plt.plot(x, w, label="w")
plt.legend()
```

```python
# 複数のグラフを別々に書く
plt.figure(figsize=(12, 4)) #横に12, 縦に4の空間に描画
# 1列目を描画(1行3列の1番目)
plt.subplot(1, 3, 1)
plt.plot(x, y, label="y")
plt.legend()
# 2列目を描画
plt.subplot(1, 3, 2)
plt.plot(x, w, label="w")
plt.legend()
# 3列目を描画
plt.subplot(1, 3, 3)
plt.plot(x, z, label="z")
plt.legend()
#グラフの重なりを防ぐ
plt.tight_layout()
```

```python
# オブジェクト指向インターフェースを用いた場合
fig, ax = plt.subplots(1, 3, figsize=(14, 4))
ax[0].plot(x, y, label="y")
ax[1].plot(x, z, label="z")
ax[2].plot(x, w, label="w")
# axに書くときはset_を頭につける必要がある
for i in range(3):
    ax[i].legend()
    ax[i].set_xlabel("x")
    ax[i].set_ylabel('y')
plt.tight_layout()
```

```python
# オブジェクト指向インターフェースを用いた場合
fig, ax = plt.subplots(2, 2)
ax[0][0].plot(x, y, label="y")
ax[0][1].plot(x, z, label="z")
ax[1][0].plot(x, w, label="w")
# axに書くときはset_を頭につける必要がある
for i in range(2):
    for j in range(2):
        ax[i][j].legend()
        ax[i][j].set_xlabel("x")
        ax[i][j].set_ylabel('y')
plt.tight_layout()
# axはnumpy配列なので[0, 0]でも良い
```

```python
# 演習2
fig, ax = plt.subplots(3, 1, figsize=(8, 6))
x = np.linspace(-np.pi, np.pi)
ax[0].plot(x, np.sin(x), label="y=sinx", color="red")
ax[1].plot(x, np.cos(x), label="y=cosx", color="blue")
ax[2].plot(x, np.tan(x), label="y=tanx", color="green")
# axに書くときはset_を頭につける必要がある
for i in range(3):
    ax[i].legend()
    ax[i].set_xlabel("x")
    ax[i].set_ylabel('y')
plt.tight_layout()
```

```python
tips = pd.read_csv('tips.csv')
tips.head(5)
```

```python
tips.isnull().sum()
```

```python
# 棒グラフ
left = range(4) #各棒のx軸上の数値
day = tips.day.value_counts().keys() #day列の要素を取り出す
height = tips.day.value_counts().values #day列の要素に対する値を取り出す
plt.bar(left, height, tick_label=day) #tick_labelはleftの値をdayの値で置き換えている
```

```python
# 円グラフ
plt.pie(tips.day.value_counts(), autopct="%.1f%%", labels=day)
plt.show()
# autopctは割合を〜％としてグラフに表示させ、3fとすれば少数第三位まで表示
```

```python
plt.hist(tips.total_bill, bins=20, color="green") #binは縦棒の数
plt.xlabel("total_bill")
plt.ylabel("counts")
plt.show()
```

```python
plt.scatter(x=tips["total_bill"], y=tips["tip"])
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.show()
```

```python
# 演習3
fig, ax = plt.subplots(2, 1, figsize=(6, 6))
man = tips[tips["sex"] == "Male"]
woman = tips[tips["sex"] != "Male"]
ax[0].scatter(x=man["total_bill"], y=man["tip"], color="blue", label="Male")
ax[1].scatter(x=man["total_bill"], y=man["tip"], color="red", label="Female")
for i in range(2):
    ax[i].set_xlabel("total_bill")
    ax[i].set_ylabel("tip")
    ax[i].legend()
```
