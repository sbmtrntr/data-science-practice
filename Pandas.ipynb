### 行と列を同時に指定して特定の値を取り出す
- df.loc[行, 列]

```python
df.loc['Goro', 'Age']
```

```python
# df[['Age', 'Sex']]と同じ処理をlocを使って書く
df.loc[:, ['Age', 'Sex']]
```

```python
# 条件を指定して取り出す
df[df['Age'] > 12]
```

```python
# 同様の操作をlocでも行える
df.loc[df['Age'] > 12, ["Height", "Weight"]]
```

```python
# データを並べ替えて出力する
df.sort_values(['Sex', 'Age'])
# 降順にするにはascending=Falseにする
# df.sort_values('Age', ascending=False)
```

```python
# Sex列に存在するユニークな値とその出現回数を求める
df["Sex"].value_counts()
```

```python
# 性別毎に集計をして、各列の平均を求める
df.groupby("Sex").mean()
```

## 欠損値の処理
1. 欠損値を含む行を削除する
2. 欠損値を他の値で置き換える

```python
#欠損したデータフレーム
df = pd.DataFrame({'Age':[12,14,12,18],
                   'Gender':['M','M','F','F'],
                   'Height':[np.nan,160,150,175], 
                   'Weight': [35,50,np.nan,60]})
df
```

```python
# 欠損値の確認
df.isnull()
```

```python
df.isnull().sum()
```

```python
# 欠損値の削除
df_dropna = df.dropna()
df_dropna
```

```python
#fillna関数を用いて平均値で欠損値を置き換え
df_fillna = df.fillna(df.mean())
df_fillna
```
