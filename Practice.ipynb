## 総合演習

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 演習1
df = pd.read_csv("house_data.csv")
plt.scatter(df["price"]*100, df["sqft_living"], s=1, color="green")
    
```

```python
# 演習2
df["good/bad"] = 0
df.loc[df["condition"] < 3, "good/bad"] = "bad"
df.loc[df["condition"] >= 3, "good/bad"] = "good"
df["good/bad"].value_counts()
```

```python
# 演習3
data = df[(df["grade"] > 3) & (df["grade"] < 6)]
sns.boxplot(data["grade"], df["price"]*100, hue=df["good/bad"])
```
