```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

## SeabornはMatplotlibをベースとして作られた可視化用ライブラリ

```python
# 棒グラフ
tips = pd.read_csv('tips.csv')
sns.countplot(tips.day)
```

```python
# 平均値の棒グラフ
# 黒の縦棒は標準偏差を表す
sns.barplot(tips.day, tips.total_bill)
```

```python
# ヒストグラム
sns.distplot(tips.total_bill, bins=20)
# kde=Falseとすると曲線なし
# hist=Falseとするとヒストグラムなし
```

```python
# 線形回帰プロット
sns.regplot(tips.total_bill, tips.tip)
# 軸ラベルが自動で付与される
# 薄い水色の領域は95%信頼区間
# scatter=Falseにすると散布図が消える
# fit_reg=Falseにすると直線が消える
```

```python
# グラフを横に並べたいときは、axにaxesを代入する
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.countplot(tips.day, ax=ax[0])
sns.regplot(tips.total_bill, tips.tip, ax=ax[1])
```

```python
# 演習1
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
sns.countplot(tips.smoker, ax=ax[0])
sns.distplot(tips.tip, kde=False, ax=ax[1])
sns.regplot(tips.total_bill, tips.tip, fit_reg=False, ax=ax[2])
```

```python
# 散布図行列
sns.pairplot(tips)
# 全ての量的変数に対して、ヒストグラムを散布図を表示
```

```python
# ヒートマップ
sns.heatmap(tips.corr(), cmap="summer_r")
```

```python
# カテゴリごとにデータの分布を表示する
sns.swarmplot("day", "total_bill", data=tips)
```

```python
sns.swarmplot("day", "total_bill", data=tips, hue="sex", dodge=True)
# hueで可視化の切り口をさらに追加
# dodge=Trueで、追加した切り口を重ねない
```

```python
sns.boxplot(tips.day, tips.tip, hue=tips.sex)
# 黒点は外れ値
```

```python
# 演習2
temp = tips[["total_bill", "tip"]]
sns.pairplot(temp)
```

```python
sns.boxplot(tips["sex"], tips["tip"], hue=tips["day"])
```
